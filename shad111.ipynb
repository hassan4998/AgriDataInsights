```python
visibility_data = df[visibility_columns]

# Calculate the Visibility Index
num_visibility_vars = len(visibility_columns)
df['Visibility Index'] = visibility_data.sum(axis=1) / num_visibility_vars

# Create Visibility Index_2: Categorize the index into 0 (Low) and 1 (High)
df['Visibility Index_2'] = df['Visibility Index'].apply(lambda x: 0 if x <= 0.5 else 1)

# Map Visibility Index_2 to categorical labels
visibility_index_2_mapping = {0: 'Low', 1: 'High'}
df['Visibility Index_2'] = df['Visibility Index_2'].map(visibility_index_2_mapping)

# Create Visibility Index_3: Categorize the index into 1 (Low), 2 (Moderate), and 3 (High)
def categorize_visibility_index_3(x):
    if x <= 0.33:
        return 1
    elif x <= 0.66:
        return 2
    else:
        return 3

df['Visibility Index_3'] = df['Visibility Index'].apply(categorize_visibility_index_3)

# Map Visibility Index_3 to categorical labels
visibility_index_3_mapping = {1: 'Low', 2: 'Moderate', 3: 'High'}
df['Visibility Index_3'] = df['Visibility Index_3'].map(visibility_index_3_mapping)

# Display the first few rows to show the new variables
df[['Visibility Index', 'Visibility Index_2', 'Visibility Index_3']].head()
```
